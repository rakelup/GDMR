```python
import pandas as pd
```

Le **DataFrame**, un tableau relationnel de données, avec des lignes et des colonnes étiquetées. Le DataFrame est une abstraction fréquemment utilisée pour manipuler des données. Spark et R proposent des implémentations similaires.

La **Series**, constituée d'une seule colonne. Un DataFrame contient une ou plusieurs Series, chacune étant étiquetée.


```python

city_names = pd.Series(['San Francisco', 'San Jose', 'Sacramento', 'New York', "Washigton"])
population = pd.Series([852469, 1015785, 485199])


df_villes = pd.DataFrame({ 'City name': city_names, 'Population': population })

df_villes['City name']
df_villes.head()

```

Acces aux élements

```python
df_villes.iloc[0,0] # Par position

print(df_villes.iloc[0]) # Par files

print(df_villes.iloc[1])

df_villes['Population'] #Par colonne

```
